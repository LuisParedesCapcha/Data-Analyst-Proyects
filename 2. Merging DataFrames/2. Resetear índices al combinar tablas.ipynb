### ¿Para qué sirvía el append? Para juntar dos datasets

```python
import pandas

sets = pandas.read_csv('MATERIALES-CAP2/02-merging-dataframes/lego-database/sets/star-wars-sets.csv')
```

```python
sets_2021 = pandas.read_csv('MATERIALES-CAP2/02-merging-dataframes/lego-database/sets/star-wars-sets_2021.csv')
```

```python
#Ahora lo uniré:
total_sets = sets.append(sets_2021)
total_sets
```

#### Fíjate en la enumeración de la columna izquierda que no posee nombre, se repiten!

```python
#Observemos las repeticiones de los índices:
total_sets.tail(10)
```

```python
#tendremos problemas localizanco al índice 0:
total_sets.loc[0]
```

### Lo arreglamos: Reseteando índices y eliminando los índices originales

```python
#ARREGLAMOS: Reseteando los índices
total_sets.reset_index()
```

```python
# Ahora eliminamos la columna "INDEX", que son los índices originales:
total_sets.reset_index(drop=True)  #Paso completo.
```
