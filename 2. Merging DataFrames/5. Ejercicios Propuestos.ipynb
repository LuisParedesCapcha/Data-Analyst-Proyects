#### Lo unimos: append

```python
#files es todos los archivos de sets-year
files = sorted(os.listdir('MATERIALES-CAP2/02-merging-dataframes/sets-year'))

#crearé un dataframa para unir todos los archivos ahí:
lego_sets = pandas.DataFrame()


#file es solo un archivo de files
for file in files: #cada uno de los archivos
    
    #leo el archivo de un año.
    lego_year = pandas.read_csv(f'MATERIALES-CAP2/02-merging-dataframes/sets-year/{file}')
    #lo uno todo en este archivo, ignorando el índice:
    lego_sets = lego_sets.append(lego_year, ignore_index = True)
```

```python
#vemos que presenta toda la información:
lego_sets
```

#### Otro modo, concat

```python
files = sorted(os.listdir('MATERIALES-CAP2/02-merging-dataframes/sets-year'))


#dataframes
dfs = [pandas.read_csv(f'MATERIALES-CAP2/02-merging-dataframes/sets-year/{file}') for file in files]
```
