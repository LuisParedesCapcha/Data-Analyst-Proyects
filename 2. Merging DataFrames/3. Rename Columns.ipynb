#### Analizando ambos datasets, difieren en la columna: theme y theme_name, que básicamente son lo mismo. Si quisiéramos unirlos con el Append, habría un problema, observa:

```python
sets.append(sets_2021)
```

#### Si observas, prevalecen ambas columnas por separado, donde aparecen valores NaN.

### Para mejorar esta situación, lo que queda sería cambiar el nombre de las columnas y poder juntarlas.

```python
#Cambiamos THEME_NAME por THEME
sets_2021 = sets_2021.rename(columns = {'theme_name': 'theme'})
sets_2021
```

```python
# Ahora sí juntaré las bases de datos, solo basta con que las columnas tengan los mismos nombres:
sets.append(sets_2021, ignore_index=True)
```
