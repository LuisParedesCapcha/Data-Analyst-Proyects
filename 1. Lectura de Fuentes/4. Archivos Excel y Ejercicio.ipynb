# Ejercicio
### A partir de los archivos pokedes de cada generación en la carpeta Materiales, generar un único libro de Excel llamado: *01-ejercicio.pokedex.xlsx* con las columnas Name, Type 1 y Type 2 en pestañas distintas para cada generación.

### Para escribir en distintas pestañas de un archivo de Excel, utilizar la clase *pandas.ExcelWriter* : https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.ExcelWriter.html

```python
import pandas
Data = pandas.ExcelFile('materiales/01-lectura-fuentes-datos/input/02-pokedex.xlsx')
Data
```

```python
pandas.read_excel(Data, usecols = ['Name', 'Type 1', 'Type 2'], sheet_name='gen_1').head()
```

```python
#Esta es la fusión pero sale el resultado feo, tipo Diccionario
Fusión = pandas.read_excel(Data, usecols = ['Name', 'Type 1', 'Type 2'], sheet_name=None)
type(Fusión)
```

```python
#Ignorando el Indexado, sí sale tipo Dataframe
Fusión = pandas.concat(pandas.read_excel(Data, usecols = ['Name', 'Type 1', 'Type 2'], sheet_name=None), ignore_index=True)
Fusión
```

```python
type(Fusión)
```

```python
import os
os.mkdir('4.EjercicioUnion')
```

```python
#El index es para borrar la columna enumarada sin nombre
Fusión.to_excel('4.EjercicioUnion/BaseUnida.xlsx', index=False)
```
